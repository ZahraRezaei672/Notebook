

# Git
Actually, it is a version control system (VCS) that controls your versions.
All VCS have a remote.
GitHub is the most famous remote for Git.

The terminal that is installed on the system after installing Git and we write Git commands in it is called Bash.

#### pwd
Print working directory.









### ls
The contents of the address you are on.







### cd
change directory.







### ls-a
./   : where we are.

../  : Previous directory.

Show hidden files.





### git
Everything we want to do with Git, we must first write git.
When we write the following command, a directory called **.git** will be created, which is not shown by ls.This command can only be used once.
Now our directory has become a repository and the project is controlled by git.







#### git status
Shows the status of the repository.








### mkdir
Create directory.






#### touch
Create file.







### rm
Delete file.


#### rm -r name directory
Delete directory.

### mv
To move or rename file and directory.
#### branch
To manage different parts of the project, we define a branch. Finally, the project manager coordinates all approved branches.
#### master
main branch.

#### git add -A
Add all changes.

#### git commit -m 'Message'
Version update.

It contains all the information that we don't want to be pushed into GitHub.
