#### If Conditions
We use if for condition.
syntax :
```python
 if condition:
      statement1
      statement2
      statement3
      .
      . 
      . 
```
If the condition is true, the statements (whatever the Python interpreter can execute) are executed.

The distance between **if** and the **condition** is a certain number of spaces.(Generally, it is 4 spaces or a tab)

If we write the first line of the statement after any number of spaces, we must put the same number of spaces for the rest of the statements so that they are placed in a block **if**.

When we want to write a statement if the condition is false, we must use **else** or **elif**.

**elif** means else if.

Each if block starts with **if** and can have any number of **elif** and can have or not have an **else**.

In each block, when the condition is true, the statement is executed, but the elifs are not executed and the block is exited.

```python
x = 5
y = 5

if x > y:
    print('x is greater that y')
elif x < y:
    print('x is smaller that y')
else:
    print('x is equal to y')
```
the output of this code :
>x is equal to y

#### | , &

- | : It means **or**.If used in the **if** condition and one of the conditions is **true**, the result is **true** and the **if** statements will be executed.
- & : It means **and**.If used in the **if** condition and one of the conditions is **false**, the result is **false** and the if statements will not be executed.

Write a program that takes a sentence as input and gives the following changes in all words
and finally print the sentence:

 The first letter of all words should move to the end.

 Add two letters ay to the end of each word.

 The first letter of each word should be capitalized.





```python
sentence = 'Python is the best programming language'

words = sentence.split()
# words = ['Python', 'is', 'the', 'best', 'programming', 'language']

new_words = []
id_ = 0
while id_ < len(words):
    # word = '' + words[id_][1:] + words[id_][0] + 'ay'
    word = ''
    word = word + words[id_][1:]
    word = word + words[id_][0]
    word = word + 'ay'
    new_words.append(word.capitalize())
    id_ = id_ + 1

new_sentence = ' '.join(new_words)
print('new_sentence:', new_sentence)
```

A function can return data as a result.

```python
def plus(a, b):
    result = a + b
    return result


# Calling  
r = plus(4, 6)
print(r)  # output 10
    
```

Some functions are internal and we can use them without defining them. Like **Sum()** , **abs()**.
```python
l1 = [1, 5, 8, 9, 7, 1, 3, 5]
total = sum(l1)
print('total:', total) # --> 39
```
But we can define it ourselves:
```python
def mysum(l1):
    '''
    arg
    out
    desc
    '''
    total = 0
    for i in l1:
        total += i
    return total


# calling
sum_ = mysum(l1)
print('sum_:', sum_)
```


### Library
Python's internal functions are of two types:
1. Standard functions (sum, abs, etc.)
2. Functions that we must import from Python libraries.
