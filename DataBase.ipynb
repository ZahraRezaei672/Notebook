








# DataBase

##### Data Base Management System(DBMS) :
In fact, it is the interface between us and the database.

Types of databases :

1.**Renational DBMS (SQL)**

2.**Non_Renational DB (NO_SQL)**


##### Data Base is a table :
- Each row of the table is a record.
- Each table has a header record that contains models or columns. (column head)
- Each column is considered a field.
- Each record contains value cells.
- In all tables, there is an ID model whose value is unique for each record.


**Note :** If we delete a record, its ID will not be deleted. For example, if we delete the last record and add a new record, the new record will be saved with the new ID.

#### Renational DB :

### Types of keys in SQL
All models can be keys, but the type is important.

#### 1. PrimeryKey(pk)
It can access a single record by itself. Each table has only one PK. It is unique and repeated only once, such as national code and student code. In many cases, ID is considered as PK.
#### 2. ForeignKey
Primarykey of one table can be used in another table.
#### 3. CompositKey
Unique access to a record by two fields or models.
#### 4. SuperKey
Unique access to a record by one or two fields or models.
#### 5. AlternateKey
It has all the characteristics of PK, but it is not PK.

### How to use foreign key
```mysql
-- Create Database
CREATE DATABASE YodaShop;

-- Use a Database
USE YodaShop;

-- Create Customers Table
CREATE TABLE Customers(
	Customer_ID INT NOT NULL,
    First_Name VARCHAR(20),
    Last_Name VARCHAR(20),
    PRIMARY KEY(Customer_ID)
);

-- Create Orders table
CREATE TABLE Orders(
	Order_ID INT NOT NULL,
    Product VARCHAR(30),
    Customer_ID INT,
    PRIMARY KEY (Order_ID),
    FOREIGN KEY (Customer_ID)
		REFERENCES Customers(Customer_ID)
);
```

### WHERE
It filters the table according to the condition written in front of it.
```mysql
SELECT * FROM INFO
WHERE Score >= 70; 

SELECT STU_No, Last_Name, Score FROM Info
WHERE Score < 30;

-- >
-- <
-- >=
-- <=
-- =
-- !=

SELECT * FROM Info
WHERE Score = 100;

SELECT Last_Name, Age FROM Info
WHERE Score != 100;
```

### ORDER BY
Sorting the table data based on the column we write in front of it and determining whether it is ascending or descending.

Examples:

```mysql
SELECT * FROM Info
ORDER BY Age;
-- ORDER BY Age ASC;

SELECT * FROM Info
ORDER BY Score DESC;

SELECT STU_No ,First_Name, Last_Name, Age FROM Info
WHERE Age > 35
ORDER BY Age ASC;
```

#### Selection Operator
Mathematical operations can also be used inside the selection.
```mysql
SELECT First_Name, Last_Name, Score, Score*10 
FROM Info;
```

A column is added to it as score*10. But this column is not added to the database. It can be named:
```mysql
SELECT First_Name,
		Last_Name,
        Score*10 AS Final_Score
FROM Info;
```

#### ROUND(x,a)
Rounds the number x to a decimal places:
```mysql
SELECT First_Name , ROUND(Score/5,2) AS FINAL FROM Info;
```

#### AND & OR
**And** : all terms together point to a record.

**Or**: Each part refers to a record.

If **and** and **or** are in the same condition, and is executed first and then its answer is executed with or.

```mysql
SELECT * FROM Info
WHERE Score > 8 AND Score < 30 AND Age > 21;

```

#### IN
It is actually like or.

```mysql
SELECT * FROM Info
WHERE First_Name IN ('Zahra', 'Marziye');
-- Where First_Name = 'Zahra' OR First_Name = 'Marziye'
```

#### BETWEEN
It is actually like and.

```mysql
SELECT * FROM Info
WHERE Score BETWEEN 70 AND 90;
-- WHERE Score >= 70 AND Score <= 90;
```

#### LIKE
A pattern for strings.
```mysql
SELECT * FROM Info
WHERE Last_Name LIKE '%ri'; -- strings that endswith 'ri'.
```

#### IS
```mysql
SELECT * From Info
WHERE Age IS NULL;

SELECT * From Info
WHERE Age IS NOT NULL;
```
#### Multiple sorting
```mysql
SELECT * FROM Info
WHERE Score >= 60
ORDER BY Score , Age DESC;
```

#### LIMIT
It displays n records that are obtained from this condition.
```mysql
SELECT * FROM Info
WHERE Score IS NOT NULL
ORDER BY Score DESC
LIMIT 5;
```

It ignores the first 5 riffs and shows the next 3 rows:

```mysql
SELECT * FROM Info
WHERE Score IS NOT NULL
ORDER BY Score DESC 
LIMIT 5, 3;
```
